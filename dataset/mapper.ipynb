```python
import csv
dataset_name = "digi_music.csv"

user_id = []
item_id = []
rating = []
timestamp = []
# Read the dataset
with open(dataset_name, 'r') as f: 
    f.readline()
    for line in f:
        line = line.strip().split(",")
        user_id.append(line[0])
        item_id.append(line[1])
        rating.append(line[2])
        timestamp.append(line[3])

unique_user_id = list(set(user_id)) 
user_id2idx = {old: new for new, old in enumerate(unique_user_id)}
unique_item_id = list(set(item_id))
item_id2idx = {old: new for new, old in enumerate(unique_item_id)}

# Write the dataset
with open("digi_music.csv", 'w') as f:
    f.write("user_id,item_id,rating,timestamp\n")
    for i in range(len(user_id)):
        f.write("{},{},{},{}\n".format(user_id2idx[user_id[i]], item_id2idx[item_id[i]], rating[i], timestamp[i]))


    
```
